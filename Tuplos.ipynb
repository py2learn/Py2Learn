<h1>Tuplo (tuple)</h1>

É um tipo de coleção ordenada que não pode ser editável.

Os tuplos em Python são escritos com parênteses simples("()")

<h4>Exemplo 1.</h4>

```python
myTuple = ("left", "right","up", "down")
```

<h3>Formas de acessar os tuplos:</h3>

Os dados dos tuplos podem ser acessados da mesma forma que nas listas, utilizando os parênteses retos e o índice desejado (positivo ou negativo).

<h3>Alterar tuplos?</h3>

Como um tuplo é uma coleção não editável, não é possível alterar os conteúdos dele e nem mesmo remover items, a não ser que, se converta o mesmo para uma lista e, utilizando os métodos da mesma se proceda a essa alteração.


<h3>Criar tuplos que contenham só um item:</h3>

Um tuplo com um item para ser considerado como sendo um, deve conter após o valor do item uma vírgula, senão não é considerado como sendo tuplo, mas sim uma <b>string</b>.

<h4>Exemplo 2.</h4>

```python
tuplo = ("koala",)
tuplo2 = ("panda")

print(type(tuplo))
print(type(tuplo2))
```

<h3>Criar tuplos com Construtor</h3>

Para criar um tuplo através do <b>construtor</b> , utilizamos o método <mark>tuple()</mark>.

<h4>Exemplo 3.:</h4>

```python
drinkTuple = tuple(("milk","coffer","juice","water"))
print(drinkTuple)
```
