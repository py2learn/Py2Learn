<h2 id="C7">Operadores Bit a bit</h2>


Têm o objetivo de comparar números/variáveis binárias (constituídas por zeros e uns).




<table style="width:100%">
  <tr>
    <th>Operador</th>
    <th>Nome</th>
    <th>Descrição</th>
  </tr>
  
  <tr>
    <td>&</td>
    <td>AND</td>
    <td>Coloca cada bit a 1 se ambos os bits estiverem a 1</td>
  </tr>
  <tr>
    <td>|</td>
    <td>OR</td>
    <td>Coloca cada bit a 1 se um dos dois bits estiverem a 1</td>
  </tr>
  <tr>
    <td>^</td>
    <td>XOR</td>
    <td>Coloca cada bit a 1 se só um dos dois bits estiverem a 1</td>
  </tr>
  <tr>
    <td>~</td>
    <td>NOT</td>
    <td>Inverte todos os bits (se estiverem a 0 passa para 1 e vice-versa)</td>
  </tr>
  <tr>
    <td><<</td>
    <td>Shift Zero à Esquerda</td>
    <td>Desloca para a esquerda pressionando zeros da direita e deixando os bits mais à esquerda</td>
  </tr>
  <tr>
    <td>>></td>
    <td>Shift Assinado à Direita</td>
    <td>Desloca para a direita empurrando cópias do bit mais à esquerda da esquerda e deixe os bits mais à direita caírem

</td>
  </tr>    
</table>
