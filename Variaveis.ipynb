<h1>Variáveis em Python</h1>


<p>Diferentemente das outras linguagens de programação, Python não tem nenhum comando específico para declarar uma variável.</p>


<h2>Regras para criar variáveis em Python</h2>


Para criar uma variável em Python é preciso ter em atenção para que:
<ul>
    <li>o nome da variável comece por uma letra ou um underscore ( _ ) e nunca por um número;</li>
    <li> o nome da variável só pode conter valores alfa-numéricos e underscores (A-z, 0-9 e _);</li>
<li> nomes das variáveis são case-sensitive (as maiúsculas e minúsculas fazem diferença, ou seja LOL não é igual a lol).</li>
</ul>
<br>

<h2> Imprimir as variáveis</h2>

Em python utiliza-se o método <mark><b>print</b>("variável")</mark> para poder imprimir o valor das variáveis.

Para combinar o texto com o valor das variáveis utiliza-se o caracter <b>"+"</b>.

<h3>Exemplo 2:</h3>

```python
modelo = "A4 Avant"
print("O meu carro de sonho é um "+ marca + " " + modelo + " " + ano)
```

<h4>Resultado:</h4><p style="font-family:consolas;">O meu carro de sonho é um Audi A4 Avant 2019</p>

 Caso a variável em questão seja um valor numérico, se adicionar <b>"+"</b> para juntar a variável a uma frase, dará erro visto que, para os números <b>"+"</b> serve como um operador matemático.

<h3>Exemplo 3:</h3>

```python
num1 = 20
num2 = 9
soma = num1 + num2
print(num1, " + ", num2, " = ",soma)
```

<h4>Resultado:</h4><p style="font-family:consolas;">20  +  9  =  29</p>
