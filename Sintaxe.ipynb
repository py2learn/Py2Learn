<h2>1. Indentações</h2>
<hr>

Na linguagem de programação Python, algo que faz muito a diferença são as indentações, ou seja são os espaços entre a margem e o início da linha de código, porque o mesmo indica um bloco de código.


<p><h3>Exemplo 1.1:</h3></p>


```python
valor = 10
if valor > 5:
print("O valor é" , valor)
```

<br></br>
 Se correres este código verás que dá erro de sintaxe visto que, não está indentado, o que indica que estamos a <br>verificar se o valor é maior que 5 e não fazemos nada em relação a isso.

 Para que o código não dê erro, é necessário que se coloque um espaço (identação) na linha de código da função <br><b>"print"</b>.
